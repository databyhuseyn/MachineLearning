```python
car_sales = pd.read_csv("/content/car-sales-extended.csv")
car_sales
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
car_sales_cat_features = X.select_dtypes(exclude = [np.number])
car_sales_cat_features
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
transformed_X = ordinal_encoder.fit_transform(car_sales_cat_features)
transformed_X
```

```python
ordinal_encoder.categories_
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot = OneHotEncoder()
transformed_X = one_hot.fit_transform(car_sales_cat_features)
transformed_X
```

```python
transformed_X.toarray()
```

```python
from sklearn.compose import ColumnTransformer         # hazir df ustunde isleyir deyirshir

one_hot = OneHotEncoder()             # classi calishdirir
cat_features = ['Make', 'Colour']         # category features teyin edir ki, bunlari secsin
transformer = ColumnTransformer([(
      'onehot', one_hot, cat_features     #pipeline edir ki, melumatlari avtomatik cevirsin       --> birincisi adidi, ikincisi yuxaridaki variabledi ki, init elesin, ucuncusu ise hansi sutuler uzre bu emeliyyati gorecek o
)], remainder = 'passthrough')            #remainder = 'passthrough'  --> qalanlari kechir

X_train_transformed = transformer.fit_transform(X_train)            # transformer pipeline'i uzre X_train datasindan goturur datani oyrenir ve transform edir

transformed_columns = transformer.get_feature_names_out()           # bu funksiyadan (pipeline) sonra hansi column cixdisa hemin columnlarin adlarini chixarir
transformed_columns

X_train_transformed_df = pd.DataFrame(X_train_transformed, columns = transformed_columns)     # DataFrame yaradiriq ki sutun adlarini gorek
X_train_transformed_df
```

```python
X_test_transformed = transformer.transform(X_test)          #fit_transform olmaz. imenni transform olmalidi, cunki yeniden label leye biler
X_test_transformed
# transformed_columns = transformer.get_feature_names_out()     ehtiyac yoxdu cunki onsuz columnlar eynidi

X_test_transformed_df = pd.DataFrame(X_test_transformed, columns = transformed_columns)
X_test_transformed_df
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor()
reg.fit(X_train_transformed_df, y_train)
```

```python
reg.score(X_train_transformed_df, y_train)
```

```python
reg.score(X_test_transformed_df, y_test)      #0.322041602821429   score ;    data azdi deye texmin ede bilmir
```

```python
car_sales_missing = car_sales_missing.dropna(subset = ['Price'])
car_sales_missing
```

```python
from sklearn.impute import SimpleImputer

cat_imputer = SimpleImputer(strategy= 'constant', fill_value = 'Missing')
num_imputer = SimpleImputer(strategy = 'median')

cat_features = ['Make', 'Colour']
num_features = ['Doors', 'Odometer (KM)']
```

```python
X = car_sales_missing.drop(["Price"], axis = 1)
y = car_sales_missing['Price'].copy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
transformer = ColumnTransformer([
    ('cat_imputer', cat_imputer, cat_features),
    ('num_imputer', num_imputer, num_features),
    ('one_hot', one_hot, cat_features)
], remainder = 'passthrough')
```

```python
X_train_transformed = transformer.fit_transform(X_train)
transform_columns = transformer.get_feature_names_out()
```

```python
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns = transform_columns)
X_train_transformed_df
```

```python
X_train_transformed_df = X_train_transformed_df.drop(['cat_imputer__Make','cat_imputer__Colour'], axis = 1)
X_train_transformed_df
```

```python
X_test_transformed = transformer.transform(X_test)
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns = transform_columns)
X_test_transformed_df
```

```python
# Step 1 : Import necessary libraries
from sklearn.pipeline import Pipeline

# Step 2 : Create Pipelines to impute and fill data
cat_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder())
])

num_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')
)])

# Step 3 : Create ColumnTransformer to applyy transformations on columns
transformer = ColumnTransformer([
    ('cat', cat_transformer, cat_features),
    ('num', num_transformer, num_features)
], remainder = 'passthrough')

# Step 4 : Fit transformations on train data
X_train_transformed = transformer.fit_transform(X_train)

# Step 5 : Create DataFrame to view data(optional)
transformed_columns = transformer.get_feature_names_out()
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns = transformed_columns)
X_train_transformed_df
```
