```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read data with pandas

housing = pd.read_csv('/content/housing.csv')
housing
```

```python
# Create new columns -- Feature Engineering

housing['rooms_per_house'] = housing['total_rooms'] / housing['households']
housing['people_per_house'] = housing['population'] / housing['households']
housing['bedrooms_ratio'] = housing['total_bedrooms'] / housing['total_rooms']
housing
```

```python
# Check Columns DataTypes

housing.info()
```

```python
# Decribe DF

housing.describe()
```

```python
# Checking if there're null values

housing.isna().sum()
```

```python
# Drop NA values in total_bedrooms column

# housing.dropna(subset = 'total_bedrooms')
```

```python
# Split Data into X and y tables

X = housing.drop('median_house_value', axis = 1)
y = housing['median_house_value'].copy()
```

```python
# Import train_test_split from Sklearn's model_selection module

from sklearn.model_selection import train_test_split
```

```python
# Split data into Train and Test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Import necessary libraries

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
# Get categorical and numerical features

cat_features = X.select_dtypes(exclude = [np.number]).columns
num_features = X.select_dtypes(include = [np.number]).columns
```

```python
# Create categorical and numeric Pipelines

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('one_hot', OneHotEncoder())
])

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('scaler', StandardScaler())
])
```

```python
# Create Transformer

transformer = ColumnTransformer([
    ('cat', cat_pipeline, cat_features),
    ('num', num_pipeline, num_features)
], remainder = 'passthrough')
```

```python
# Create Full Pipeline to implement all steps

full_pipeline_regression = Pipeline([
    ('preprocessing', transformer),
    ('regression', RandomForestRegressor())
])

# Fit Model

full_pipeline_regression.fit(X_train, y_train)
```

```python
# Check Train set score

full_pipeline_regression.score(X_train,y_train)
```

```python
# Check Test set score

full_pipeline_regression.score(X_test,y_test)
```

```python
# Import mean_absolute_error from Sklearn's Metrics module

from sklearn.metrics import mean_absolute_error
```

```python
# Get Model's prediction on X_test dataset

reg_preds = full_pipeline_regression.predict(X_test)
```

```python
# Calculate mean_absolute_error

mean_absolute_error(y_test, reg_preds)
```

```python
# Import cross_val_score from Sklearn's model_selection module

from sklearn.model_selection import cross_val_score


# Calculate Cross validation score
cross_val_score(full_pipeline_regression, X_test, y_test, cv = 10).min()
```

```python
# Import GridSearchCV from Sklearn's model_selection module

from sklearn.model_selection import GridSearchCV
```

```python
# Create parameter grid

param_grid_reg = {
    'regression__n_estimators' : [10,20,30],
    'regression__criterion' : ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']
}

# Implement GridSearchCV on RandomForestRegressor
grid_search_random_forest_reg = GridSearchCV(full_pipeline_regression, param_grid_reg, cv = 3)
grid_search_random_forest_reg.fit(X_train, y_train)
```

```python
# Check best parameter

grid_search_random_forest_reg.best_params_
```
