```python
# Import necessary libraries

import pandas as pd     # For data manipulation and analysis
import numpy as np      # For numerical operations
import matplotlib.pyplot as plt   # For data visualization
```

```python
# Read the main dataset from a CSV file into a DataFrame
main_dataset = pd.read_csv('/content/train.csv')

# Read the test dataset from a CSV file into a DataFrame
test_dataset = pd.read_csv('/content/test.csv')

# Set the display option to show all columns when displaying the DataFrame
pd.set_option('display.max_columns', None)
```

```python
# Display information about the main dataset
main_dataset.info()
```

```python
# Count the number of missing values in each column of the main dataset
main_dataset.isna().sum()
```

```python
# Drop unnecessary columns from the main dataset
main_dataset = main_dataset.drop(['ID','Customer_ID', 'SSN', 'Name'], axis=1)
```

```python
# Drop NA values if there's in Target Column
main_dataset.dropna(subset = 'Credit_Score', inplace = True)
```

```python
# Replace underscore's to turn some category columns into numeric columns

main_dataset['Age'] = main_dataset['Age'].str.replace('_','')
main_dataset['Annual_Income'] = main_dataset['Annual_Income'].str.replace('_','')
main_dataset['Num_of_Loan'] = main_dataset['Num_of_Loan'].str.replace('_','')
main_dataset['Num_of_Delayed_Payment'] = main_dataset['Num_of_Delayed_Payment'].str.replace('_','')
main_dataset['Changed_Credit_Limit'] = main_dataset['Changed_Credit_Limit'].str.replace('_','')
main_dataset['Outstanding_Debt'] = main_dataset['Outstanding_Debt'].str.replace('_','')
main_dataset['Amount_invested_monthly'] = main_dataset['Amount_invested_monthly'].str.replace('_','')
main_dataset['Monthly_Balance'] = main_dataset['Monthly_Balance'].str.replace('_','')
```

```python
# Turn these columns into numeric

main_dataset[['Age',
       'Annual_Income',
       'Num_of_Loan',
       'Num_of_Delayed_Payment',
       'Changed_Credit_Limit',
       'Outstanding_Debt',
       'Amount_invested_monthly',
       'Monthly_Balance']] = main_dataset[['Age',
                                    'Annual_Income',
                                    'Num_of_Loan',
                                    'Num_of_Delayed_Payment',
                                    'Changed_Credit_Limit',
                                    'Outstanding_Debt',
                                    'Amount_invested_monthly',
                                    'Monthly_Balance']].apply(pd.to_numeric,
                                                              errors='coerce')
```

```python
# Turn Credit_History_Age column into date format

main_dataset['Credit_History_Age'] = pd.to_datetime(main_dataset['Credit_History_Age'], format='%Y Years and %m Months', errors='coerce')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Clean and preprocess the 'Type_of_Loan' column
main_dataset['Type_of_Loan'] = main_dataset['Type_of_Loan'].str.replace('and','').str.replace(' ','').str.replace(' and','').str.replace('and ','')
main_dataset['Type_of_Loan'] = main_dataset['Type_of_Loan'].astype(str)

# Split the 'Type_of_Loan' values into lists and replace certain strings
loan_type_array = main_dataset['Type_of_Loan'].str.split(',').replace('and','').replace(' ','').replace(' and','').replace('and ','')

# Initialize MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Encode the 'Type_of_Loan' values into binary columns
encoded_loans = pd.DataFrame(mlb.fit_transform(loan_type_array), columns=mlb.classes_, index=main_dataset.index)
```

```python
# Display encoded_loans table
encoded_loans
```

```python
# Join the main dataset with the encoded loan types DataFrame
train = main_dataset.join(encoded_loans)

# Drop the original 'Type_of_Loan' column as it has been encoded
train.drop(['Type_of_Loan'], axis = 1, inplace = True)

train
```

```python
# Fix incredible Ages

train.loc[train['Age'] < 13, 'Age'] = train.loc[train['Age'] < 13, 'Age'] * (-0.1)
train.loc[train['Age'] > 99] = train.loc[train['Age'] > 99].drop('Age',axis = 1)
```

```python
# Show max and min value of Age column

train['Age'].max(), train['Age'].min()
```

```python
# Describe Train Table

train.describe()
```

```python
# Create X and y tables

X = train.drop(['Credit_Score'], axis=1)
y = train['Credit_Score'].copy()
```

```python
# Turn y series into DataFrame

y = pd.DataFrame(y)
```

```python
X
```

```python
y
```

```python
# Here we can see Imbalance in dataset. So we should use stratify.

y.value_counts()
```

```python
# Split data into train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify= y)
```

```python
# Import libraries from Scikit-Learn

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
# Get names of numeric and categorical features separately and drop Credit_Score to prevent it to be transformed

num_features = train.select_dtypes(include=[np.number]).columns
cat_features = train.select_dtypes(exclude=[np.number]).columns.drop('Credit_Score')
```

```python
# Check feature names

num_features, cat_features
```

```python
# Import Classifiers

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create Pipelines and Transformer

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy = 'median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('one_hot', OneHotEncoder())
])


transformer = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
], remainder = 'passthrough')
```

```python
# Create full pipeline for KNearestNeighbors and RandomForestClassifier

full_pipeline_knn = Pipeline([
    ('preprocessing', transformer),
    ('knn', KNeighborsClassifier())
])

full_pipeline_rfc = Pipeline([
    ('preprocessing', transformer),
    ('rfc', RandomForestClassifier())
])
```

```python
# Train the RandomForestClassifier model using the training data

full_pipeline_rfc.fit(X_train, y_train)
```

```python
# Print RandomForestClassifier training and testing set scores
print(f' RandomForestClassifier Train Set Score: {full_pipeline_rfc.score(X_train, y_train)} \n RandomForestClassifier Test Set Score: {full_pipeline_rfc.score(X_test, y_test)}')
```

```python
# Train the K-Nearest Neighbors (KNN) model using the training data
full_pipeline_knn.fit(X_train, y_train)
```

```python
# Print K-Nearest Neighbors (KNN) model training and testing set scores
print(f' KNearestNeighbors Train Set Score: {full_pipeline_knn.score(X_train, y_train)} \n KNearestNeighbors Test Set Score: {full_pipeline_knn.score(X_test, y_test)}')
```

```python
# Import GridSearchCV from scikit-learn for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid for RandomForestClassifier
param_grid_rfc = {
    'rfc__n_estimators': [50, 100, 200],
    'rfc__criterion': ['gini', 'entropy', 'log_loss']
}

# Flatten y_train (convert to 1D array)
y_train = np.ravel(y_train)

# Create GridSearchCV object for RandomForestClassifier
grid_rfc = GridSearchCV(full_pipeline_rfc, param_grid_rfc, cv=3)

# Fit the GridSearchCV object on the training data to find the best hyperparameters
grid_rfc.fit(X_train, y_train)
```

```python
# Print the best hyperparameters found during grid search for RandomForestClassifier
for parameter, value in grid_rfc.best_params_.items():
  print(f'Best value for {parameter} is {value}')
```

```python
# Fit the RandomForestClassifier with the best estimator obtained from grid search on the training data
grid_rfc.best_estimator_.fit(X_train, y_train)

# Print RandomForestClassifier training and testing set scores using the best estimator
print(f' RandomForestClassifier with best estimator Train Set Score: {grid_rfc.best_estimator_.score(X_train, y_train)} \n RandomForestClassifier with best estimator Test Set Score: {grid_rfc.best_estimator_.score(X_test, y_test)}')
```

```python
# Use the best estimator from grid search to make predictions on the test data
prediction_rfc = grid_rfc.best_estimator_.predict(X_test)
```

```python
# Import classification evaluation metrics from scikit-learn
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# - precision_score measures the ratio of correctly predicted positive observations to the total predicted positives.
# - recall_score measures the ratio of correctly predicted positive observations to the all observations in actual class.
# - f1_score is the harmonic mean of precision and recall, providing a balanced measure between the two.

rfc_precision_score = precision_score(y_test, prediction_rfc, average='macro')
rfc_recall_score = recall_score(y_test, prediction_rfc, average='macro')
rfc_f1_score = f1_score(y_test, prediction_rfc, average='macro')
```

```python
# Print precision, recall, and F1-score for RandomForestClassifier predictions
print(f'''
precision score : {rfc_precision_score}
recall score :    {rfc_recall_score}
f1 score :        {rfc_f1_score}
''')
```

```python
# Import the classification_report function from scikit-learn metrics
from sklearn.metrics import classification_report

# Print the classification report for RandomForestClassifier predictions
print(classification_report(prediction_rfc, y_test))
```

```python
# Define the parameter grid for K-Nearest Neighbors (KNN)
param_grid_knn = {
    'knn__n_neighbors': [5 ,10,15],
    'knn__leaf_size': [10, 30 ,50]
}

# Create a GridSearchCV object for K-Nearest Neighbors (KNN) using the specified parameter grid
grid_knn = GridSearchCV(full_pipeline_knn, param_grid_knn, cv=3)

# Perform grid search on the training data to find the best hyperparameters for KNN
# Subset the training data and labels for more efficient grid search (example: first 10,000 samples)
grid_knn.fit(X_train[:10000], y_train[:10000])
```

```python
# Iterate over the best hyperparameters found during grid search for KNN
for parameter, value in grid_knn.best_params_.items():
  print(f'Best value for {parameter} is {value}')
```

```python
# Fit the K-Nearest Neighbors (KNN) model with the best estimator obtained from grid search on the training data
grid_knn.best_estimator_.fit(X_train, y_train)

# Print K-Nearest Neighbors (KNN) training and testing set scores using the best estimator
print(f' KNearestNeighbors with best estimator Train Set Score: {grid_knn.best_estimator_.score(X_train, y_train)} \n KNearestNeighbors with best estimator Test Set Score: {grid_knn.best_estimator_.score(X_test, y_test)}')
```

```python
# Use the best estimator from grid search to make predictions on the test data
prediction_knn = grid_knn.best_estimator_.predict(X_test)
```

```python
# Calculate precision, recall, and F1-score for K-Nearest Neighbors (KNN) predictions
knn_precision_score = precision_score(y_test, prediction_knn, average='macro')
knn_recall_score = recall_score(y_test, prediction_knn, average='macro')
knn_f1_score = f1_score(y_test, prediction_knn, average='macro')
```

```python
# Print precision, recall, and F1-score for K-Nearest Neighbors (KNN) predictions
print(f'''
precision score : {knn_precision_score}
recall score :    {knn_recall_score}
f1 score :        {knn_f1_score}
''')
```

```python
# Print the classification report for K-Nearest Neighbors (KNN) predictions
print(classification_report(prediction_knn, y_test))
```
