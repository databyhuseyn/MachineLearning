```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_set = pd.read_csv('/content/train.csv')
test_set = pd.read_csv('/content/test.csv')
gender_submission = pd.read_csv('/content/gender_submission.csv')
```

```python
train_set = train_set.drop([61,829], axis = 0)
train_set
```

```python
X_train = train_set.drop(['Name','Survived','Ticket','Cabin'], axis = 1)
y_train = train_set['Survived']
```

```python
X_test = test_set
```

```python
X_train.set_index('PassengerId', inplace = True)
```

```python
X_train
```

```python
X_train.info()
```

```python
from sklearn.impute import SimpleImputer

cat_imputer = SimpleImputer(strategy= 'constant', fill_value = 'Missing')
num_imputer = SimpleImputer(strategy= 'median')

cat_features = ['Sex','Embarked']
num_features = ['Pclass','Age','SibSp','Parch','Fare']
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer

# categorical_values = X_train.select_dtypes(exclude = [np.number])

ordinal_encoder = OrdinalEncoder()
# X_train_transformed = one_hot.fit_transform(categorical_values)
# X_train_transformed
```

```python
transformer = ColumnTransformer([
    ('cat_imputer', cat_imputer, cat_features),
    ('num_imputer', num_imputer, num_features),
    ('ordinal_encoder', ordinal_encoder, cat_features)
], remainder = 'passthrough')
```

```python
X_train_transformed = transformer.fit_transform(X_train)
transform_columns = transformer.get_feature_names_out()
```

```python
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns = transform_columns)
X_train_transformed_df
```

```python
X_train_transformed_df = X_train_transformed_df.drop(['cat_imputer__Sex','cat_imputer__Embarked'], axis = 1)
X_train_transformed_df
```

```python
X_test_transformed = transformer.transform(X_test)
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns = transform_columns)
X_test_transformed_df = X_test_transformed_df.drop(['cat_imputer__Sex','cat_imputer__Embarked'], axis = 1)
X_test_transformed_df
```

```python
X_train_transformed_df.isna().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train_transformed_df, y_train)
```

```python
clf.score(X_train_transformed_df, y_train)
```

```python
clf.score(X_test_transformed_df,gender_submission['Survived'])
```
